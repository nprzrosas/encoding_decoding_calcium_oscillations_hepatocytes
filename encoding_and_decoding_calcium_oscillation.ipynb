### **Introduction and Background** <br />
 Cells can send and receive information by controlling the spatial and temporal behavior of signaling molecules. Different upstream signals can lead to different dynamical patterns of the same molecule. [Fig 1] In the case of Ca$^{2+}$ signaling in hepatocytes, hormones such as vasopressin and angiotensin II or nucleotides bind to membrane-spanning receptor proteins to induce signaling. <br>

<image src="https://drive.google.com/uc?id=1ClR42Pn4ZavOntweTw2Nz3gfEtLOeqD_" width="1000"> <br>

Ca$^{2+}$ is a versatile intracellular signaling molecule in many different cell types. It is a widespread messenger system in many cells. Information is transferred from extracellular agonists binding to the receptor enabling Ca$^{2+}$ release. The Ca$^{2+}$ directly activate or indirectly activate genes via Ca$^{2+}$ activated proteins. This project investigates how the cell interprets different types of information stimulated by different agonists but encoded universally by Ca$^{2+}$ signals. The hypothesis of the project is information in the Ca$^{2+}$ oscillations is encoded in its frequency, amplitude, and its waveform. Cells use various regulatory proteins which cooperatively bind to Ca$^{2+}$  to decode this complex signaling. <br>
<image src="https://drive.google.com/uc?id=1GHq66k7Utdyn2-zipxbJ2wxHBI7vKIH_" width="500"> <br>
In hepatocytes, agonists like nucleotides and vasopressin binds to the extracellular part of the G-proteins. The G$_{\alpha}$ subunit gets activated by exchanging GDP for GTP. The activated G$_{\alpha}$ subunit acts on the PLC enzyme, this in turn hydrolyses membrane bound phosphatidyl inositol 4,5-bisphosphate (PIP$_2$) to inositol 1,4,5-trisphosphate (IP$_3$) and diacylglycerol (DAG).[1] The IP$_3$ then binds to the IP$_3$R (IP$_3$ receptor) on the endoplasmic reticulum (ER) releasing the stored Ca$^{2+}$ into the cytosol. There is also release of Ca$^{2+}$ from mitochondria through the process of Ca$^{2+}$ induced Ca$^{2+}$ release (CICR). Ca$^{2+}$ also enters the cells from the extracellular space through membrane channels. The removal of Ca$^{2+}$ from the cytosol is mediated by Sarco/Endoplasmic Reticulum Ca$^{2+}$-ATPase (SERCA), an ATP dependent Ca$^{2+}$ pump on the ER and the plasma membrane Ca$^{2+}$ ATPase (PMCA) pump localized on the plasma membrane.[2] [Fig 2]. <br>

## Computational Model


The model proposed is a modification of a previously formulated receptor operated model (Kummer et al 2000).  This model investigates the cooperative Ca$^{2+}$ binding to proteins and the different activities of these proteins in response to different types of Ca$^{2+}$ oscillations. The model consists of 5 variables represented in the form of ODEs:

+ Activated (G$_{\alpha}$) subunit of the receptor complex (G$_{\alpha}$)
+ Active phospholipase C (PLC)
+ Cytosolic Ca$^{2+}$ concentrations (Ca$_{cyt}$)
+ Concentration of free Ca$^{2+}$ in the ER (Ca$_{ER}$)
+ Ca$^{2+}$ concentration in the mitochondria (Ca$_{mit}$) <br>

In this section the variables are explained.

1.  **Activated Gα subunit of the receptor complex (G$_{\alpha}$)**.
The variable indicates concentration of active G$_{\alpha}$ subunit of the G-protein. The G-protein is coupled to a receptor on the cytosolic side that binds to agonists like vasopressin and nucleotides. In this model, depending on the receptor type, the G$_{\alpha}$ subunit autocatalytically activates other Gα subunit. This is represented by the term k$_2$ × G$_{\alpha}$ (agonist specific autocatalytic activation). The spontaneous or agonist activation is represented by k$_1$. Inactivation of G$_{\alpha}$ depends on active PLC which activates a GTP hydrolysis protein thus reducing the active G$_{\alpha}$. The Ca$^{2+}$ dependent kinase also inactivates the Gα complex.

3. **Concentration of free Ca$^{2+}$ in the ER (Ca$_{ER}$)** <br>
In rat hepatocytes, Ca$^{2+}$ oscillations induced by agonist binding does not induce complete depletion of Ca$^{2+}$ in ER. The IP$_3$R channel is regulated by IP$_3$, ATP, pH, and cytosolic Ca$^{2+}$. The cooperative binding of IP$_3$ to the tetrameric channel causes the channel to open and allows the flux of Ca$^{2+}$. The Ca$^{2+}$ binds to the IP$_3$ and stabilizes the open state of the IP$_3$R (an effect called Ca$^{2+}$ Induced Ca$^{2+}$ Release). The Hill coefficient for the IP$_3$ binding to IP$_3$ receptor is 4. IP$_3$ channel in this model is considered to mediate diffusion in both directions. <br>

4. **Ca$^{2+}$ concentration in the mitochondria (Ca mit)** <br>
The Rapid Mode of Uptake Mechanism (RaM) is considered for modelling Ca$^{2+}$ uptake in mitochondria. In RaM mechanism, a significant amount of Ca$^{2+}$ is sequestered in the beginning of each cytosolic Ca$^{2+}$ pulse. Due to the switch-like uptake of Ca$^{2+}$, the Ca$^{2+}$ concentration is modelled using a Hill term with a coefficient of 8. Ca$^{2+}$ release from the mitochondria is mediated by permeability transition pore (PTP). The PTP under normal conditions exists in a closed state and low conduction state that allows limited diffusion of Ca$^{2+}$. The PTP is modeled considering the mCICR (mitochondrial Ca$^{2+}$ induced Ca$^{2+}$ release), in an open state with diffusion of Ca$^{2+}$ taking place both directions.

## Model features and assumptions <br>
Model used is qualitative. It deals with the levels of Ca$^{2+}$ in the different compartments and not concentration of the Ca$^{2+}$. It doesn’t consider the sizes of the different compartments.
+ For the PLC variable: The activation of PLC could be a Michaelis–Menten equation, but KM is very small then (k$_7$ × G$_{\alpha}$ × PLC) / (K$_M$ + PLC) = k$_7$ × G$_{\alpha}$.
+ In the Ca$_{ER}$ variable, because of the very short half-life of IP$_3$, concentration is assumed to be in quasi steady state with PLC. Therefore, PLC is used as a parameter where IP$_3$ should be used.
+ In the Ca$_{mit}$ variable, the change in the mitochondrial membrane potential is held constant and is implicitly included in k$_18$ and k$_20$. This model only considers a low conductance state for permeability transition pore (responsible for Ca$^{2+}$ release)
+ For the hypothetical enzyme system in the Task 4, it is assumed that the concentration of substrate for the reaction remains constant. The change in cytosolic Ca$^{2+}$ concentration due to binding of Ca$^{2+}$ to the enzyme is insignificant compared to the amplitude of Ca$^{2+}$ oscillations.

```python
# To setup the Bursting Model
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
import math as m

k1 = 0.01
k2 = 2.0
k3 = 0.64
K4 = 0.09
k5 = 4.88
K6 = 1.18
k7 = 2.08
k8 = 32.24
K9 = 29.09
k10 = 0.7
K11 = 3
k12 = 2.8
k13 = 13.4
k14 = 153
K15 = 0.16
k16 = 7
K17 = 0.05
k18 = 79
K19 = 3.5
k20 = 0.81
K21 = 4.5

# Initial conditions

g_alpha0 = 0.01
plc0 = 0.01
c_cyt0 = 0.01
c_ret0 = 10
c_mit0 = 0.001

# list of the initial conditions
n0 = [g_alpha0, plc0, c_cyt0, c_ret0, c_mit0]


# Time course
z = 1000000
seconds = 400
t = np.linspace(0, seconds ,z)

# create a function fo solve a set of ODE's

def calcium(n, t1):
  g_alpha = n[0]
  plc = n[1]
  c_cyt = n[2]
  c_ret = n[3]
  c_mit = n[4]

  # create the differential equations
  dg_alphadt = k1 + k2 * g_alpha - (k3 * g_alpha * plc)/(g_alpha + K4)-(k5 * c_cyt * g_alpha)/(g_alpha + K6)
  dplcdt = k7 * g_alpha - (k8 * plc) / (plc + K9)
  dc_cytdt = (c_ret-c_cyt) * (k10 * c_cyt * plc ** 4)/(plc ** 4 + K11 ** 4) + (k12 * plc) + (k13 * g_alpha) - (k14 * c_cyt)/(c_cyt +K15) - (k16 * c_cyt)/(c_cyt + K17) - (k18 * c_cyt**8)/(K19 ** 8 + c_cyt**8) + (c_mit-c_cyt) * (k20 * c_cyt)/(c_cyt + K21)
  dc_retdt = - (c_ret - c_cyt) * (k10 * c_cyt * plc ** 4)/(plc ** 4 + K11 ** 4) + (k16 * c_cyt)/(c_cyt + K17)
  dc_mitdt = (k18 * c_cyt ** 8)/(K19 ** 8 + c_cyt ** 8) - (c_mit -c_cyt) * (k20 * c_cyt)/(c_cyt + K21)


  return [dg_alphadt, dplcdt, dc_cytdt, dc_retdt, dc_mitdt]

sol_n = odeint(calcium, n0, t)

# Assign variable names to solution matrix
g_alphan = sol_n[:,0]
plcn = sol_n[:,1]
c_cytn = sol_n[:,2]
c_retn = sol_n[:,3]
c_mitn = sol_n[:,4]

plt.figure(num=34)
plt.plot(t,g_alphan, label = "G alpha")
plt.plot(t,plcn, label = "PLC")
plt.plot(t,c_cytn, label = "Ca2+ cyt")
plt.plot(t,c_retn, label = "Ca2+ ret" )
plt.plot(t,c_mitn, label = "Ca2+ mit" )

plt.xlabel("s")
plt.ylabel("Ca2+ cyt(Arbitrary Units)")
plt.legend(loc=0, bbox_to_anchor=(1.2,1))


plt.figure(num=35)
plt.plot(t,c_cytn, label = "Ca2+ cyt")
plt.title('k2 = 2.235 -> Bursting')
plt.xlabel("s")
plt.ylabel("Ca2+ cyt (Arbitrary Units)")
plt.legend(loc=0, bbox_to_anchor=(1.2,1))


plt.figure(num=36)
plt.plot(t[0:100000],c_cytn[0:100000], label = "Ca2+ cyt")
plt.title('k2 = 2.235 -> Bursting')
plt.xlabel("s")
plt.ylabel("Ca2+ cyt (Arbitrary Units)")
plt.legend(loc=0, bbox_to_anchor=(1.2,1))
```
